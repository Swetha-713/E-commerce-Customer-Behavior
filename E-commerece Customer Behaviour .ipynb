```python
# Step 4: Check for missing values
print("\nMissing values in each column:")
print(df.isnull().sum())
```

```python
# Step 5: Remove duplicates
duplicates = df.duplicated()
print("\nNumber of duplicate rows:", duplicates.sum())
df = df.drop_duplicates()
```

```python
# Step 6: Convert date columns to datetime format if any
# Example: If there is a 'Order Date' column
if 'Order Date' in df.columns:
    df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')
```

```python
# Step 7: Handling missing values
# Drop columns with too many missing values or fill with mean/median/mode
threshold = 0.4  # if more than 40% missing, drop the column
missing_ratio = df.isnull().mean()
cols_to_drop = missing_ratio[missing_ratio > threshold].index
df = df.drop(columns=cols_to_drop)

# Fill remaining missing values
for col in df.columns:
    if df[col].dtype == 'object':
        df[col].fillna(df[col].mode()[0], inplace=True)
    else:
        df[col].fillna(df[col].median(), inplace=True)
```

```python
# Step 8: Drop irrelevant columns (example: 'Customer ID', 'Unnamed: 0' if exists)
columns_to_remove = ['Customer ID', 'Unnamed: 0']
df = df.drop(columns=[col for col in columns_to_remove if col in df.columns])
```

```python
# Step 9: Rename columns to lowercase and replace spaces with underscores
df.columns = df.columns.str.lower().str.strip().str.replace(' ', '_')
```

```python
# Step 10: Save cleaned data to new CSV
df.to_csv('Cleaned_Ecommerce_Data.csv', index=False)
print("\nCleaned dataset saved as 'Cleaned_Ecommerce_Data.csv'")
```

# Charts

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\USER\Ecommerce_Consumer_Behavior_Analysis_Data.csv")
gender_counts = df['Gender'].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140, colors=['#FF9999','#66B2FF'])
plt.title('Gender Distribution')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
category_counts = df['Purchase_Category'].value_counts().head(10)

plt.figure(figsize=(10, 6))
category_counts.plot(kind='bar', color='skyblue')
plt.title('Top 10 Purchase Categories')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df['Customer_Satisfaction'] = df['Customer_Satisfaction'].astype(int)

plt.figure(figsize=(8, 5))
df['Customer_Satisfaction'].plot(kind='hist', bins=10, color='orange', edgecolor='black')
plt.title('Customer Satisfaction Distribution')
plt.xlabel('Satisfaction Score')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
device_counts = df['Device_Used_for_Shopping'].value_counts()

plt.figure(figsize=(7, 7))
plt.pie(device_counts, labels=device_counts.index, autopct='%1.1f%%', startangle=140, colors=['#FFCC99', '#99CCFF', '#FF6666'])
plt.title('Device Used for Shopping')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df.boxplot(column='Purchase_Amount', by='Income_Level', grid=False)
plt.title('Income Level vs Purchase Amount')
plt.suptitle('')
plt.xlabel('Income Level')
plt.ylabel('Purchase Amount')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
avg_purchase = df.groupby('Marital_Status')['Purchase_Amount'].mean()
plt.figure(figsize=(8, 5))
avg_purchase.plot(kind='bar', color='lightgreen')
plt.title('Average Purchase Amount by Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Average Purchase Amount')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
loyalty_data = pd.crosstab(df['Customer_Loyalty_Program_Member'], df['Purchase_Intent'])
loyalty_data.plot(kind='bar', stacked=True, figsize=(8, 6), colormap='viridis')
plt.title('Customer_Loyalty_Program_Member vs Purchase Intent')
plt.xlabel('Customer_Loyalty_Program_Member')
plt.ylabel('Number of Customers')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
